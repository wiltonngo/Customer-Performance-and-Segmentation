# 4. Aggregating Data with Transform()

```python
ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

# Deriving Columns with loc()

```python
# Create a flag for new Column for loyatly_flag
ords_prods_merged.loc[ords_prods_merged['max_order']> 40,'loyatly_flags'] = 'Loyal customer'
ords_prods_merged.loc[ords_prods_merged['max_order']<= 40 & (ords_prods_merged['max_order'] > 10),'loyatly_flags'] = 'Regular customer'
ords_prods_merged.loc[ords_prods_merged['max_order']<= 10,'loyatly_flags'] = 'New customer'
```

```python
# Check the Frequency for the Loyatly_flags Column
ords_prods_merged['loyatly_flags'].value_counts(dropna = False)
```

```python
ords_prods_merged.head()
```

```python
#Performing multiple aggregations ('min', 'max', 'mean', 'sum')
ords_prods_merged.groupby('loyatly_flags').agg({'prices': ['mean', 'min', 'max', 'sum']})
```

### Loyal customers purchase cheaper products. Where as Regular products purchase more expensive products

```python
#Split the data into groups based on the 'user_id' column. Apply the transform() function on the 'prices' column to generate the average prices for each user.
ords_prods_merged['avg_price'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merged.head()
```

```python
ords_prods_merged.loc[ords_prods_merged['avg_price'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merged.loc[ords_prods_merged['avg_price'] <= 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

```python
##Split the data into groups based on the 'user_id' column. Apply the transform() function on the 'days_since_prior_order' column to generate the median order frequency for each user.
ords_prods_merged['Meidan_Freq'] = ords_prods_merged.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
ords_prods_merged.loc[ords_prods_merged['Meidan_Freq'] > 20, 'order_freq_flag'] = 'Non Frequent Customer'
ords_prods_merged.loc[(ords_prods_merged['Meidan_Freq'] > 10) & (ords_prods_merged['Meidan_Freq'] <= 20) , 'order_freq_flag'] = 'Regular Customer'
ords_prods_merged.loc[ords_prods_merged['Meidan_Freq'] <= 10, 'order_freq_flag'] = 'Frequent Customer'
```

```python
ords_prods_merged['order_freq_flag'].value_counts()
```
