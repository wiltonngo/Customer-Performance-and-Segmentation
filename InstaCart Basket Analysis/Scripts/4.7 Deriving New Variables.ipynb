```python
# Create a subset 
df = ords_prods_merged[:1000000]
```

```python
# Create and assign these flags, and write a user-defined function
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] >5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
# Define the price_label and apply to the subset you created
df['prices_range'] = df.apply(price_label,axis=1)
```

```python
# Check the value_counts for your new Column
df['prices_range'].value_counts(dropna = False)
```

```python
# Check themost exepensive proudct of the subset is
df['prices'].max()
```

# 4. IF-Statements with For-Loops

```python
# Which day that themost orders take place
ords_prods_merged['order_day_of_week'].value_counts()
```

```python
result = []
```

```python
for value in ords_prods_merged["order_day_of_week"]:
    if value == 0:
        result.append("Busiest day")
    elif value == 4:
        result.append("Least busy")
    else:
        result.append("Regularly busy")
```

```python
result
```

```python
# Combine with your dataframe and create a new column
ords_prods_merged['busiest_day'] = result
```

```python
# Check the frequency of orders_day_of_the_week
ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

```python
# Drop the Column Busiest Day
ords_prods_merged.drop(columns = ['busiest_day'],inplace = True)
```

```python
# Create the Column that create for the busiest_ days and slowest days
Busiest_day = []

for value in ords_prods_merged["order_day_of_week"]:
    if value == 0 or value == 1:
        Busiest_day.append("Busiest days")
    elif value == 4 or value == 3:
       Busiest_day.append("Slowest Days")
    else:
        Busiest_day.append("Regularly busy")
```

```python
# Create the busiest Day colum
ords_prods_merged['Busiest Day'] = Busiest_day
```

```python
# Get the Frequency of Busiest Day column
ords_prods_merged['Busiest Day'].value_counts()
```

```python
# Identify the Busiest hour of the Day
ords_prods_merged['order_hour_of_day'].value_counts()
```

# 5. Create Busiest time Column

```python
# create a for-loop that append to the 'hours' list the busyness of the hours based on values in the 'order_hour_of_day' column
hours = []

for hour in ords_prods_merged['order_hour_of_day']:
  if hour in [10, 11, 14, 15, 13, 12, 16, 9]:
    hours.append('Most orders')
  elif hour in [23, 6, 0, 1, 5, 2, 4, 3]:
    hours.append('Fewest orders')
  else:
    hours.append('Average orders')
```

```python
# Createa a new column for busiest hours
ords_prods_merged['Busiest Period of Day'] = hours
```

```python
# Check the Frequency of Busiest Periods
ords_prods_merged['Busiest Period of Day'].value_counts()
```
